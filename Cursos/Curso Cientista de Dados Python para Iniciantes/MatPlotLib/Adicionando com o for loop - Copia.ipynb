```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('Tabelas/gas.csv')
gas
```

```python
plt.title('Valores do Gas Natural em US$ / Anos 90 -2008')

list_country = ['Australia' , 'USA', 'Italy', 'Mexico', 'France', 'Germany']

for country in gas:
    if country in list_country:
        plt.plot(gas['Year'], gas[country], label = [country], marker ='.')

plt.xticks(gas['Year'][::2]) #Escala X de 2 em 2 anos 

plt.xlabel('Years') #leganda em x
plt.ylabel('US$') #legenda em Y

plt.legend()
plt.show()
```
