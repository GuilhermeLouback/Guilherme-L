```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('Tabelas/gas.csv')
gas
```

```python
plt.title('Valores do Gas Natural em US$ / Anos 90 -2008')

plt.plot(gas['Year'], gas['Australia'],'r.--' ,label = 'Australia') #label cria legenda 
plt.plot(gas['Year'], gas['Italy'],'g' ,label = 'Itália')
plt.plot(gas['Year'], gas['USA'], 'b.-',label = 'Estados Unidos') #alterar cores e tipo de linha antes de label
plt.plot(gas['Year'], gas['Japan'], 'y',label = 'Japão')


plt.xticks(gas['Year'][::2]) #Escala X de 2 em 2 anos 

plt.xlabel('Years') #leganda em x
plt.ylabel('US$') #legenda em Y

plt.legend()
plt.show()
```
