```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gas = pd.read_csv('Tabelas/gas.csv')
gas
```

```python
plt.title('Valores do Gas Natural em US$ / Anos 90 -2008')

plt.plot(gas['Year'], gas['Australia'], label = 'Australia') #label cria legenda 
plt.plot(gas['Year'], gas['Italy'], label = 'Itália')
plt.plot(gas['Year'], gas['USA'], label = 'Estados Unidos')
plt.plot(gas['Year'], gas['Japan'], label = 'Japão')


plt.xticks(gas['Year'][::2]) #Escala X de 2 em 2 anos 

plt.legend()
plt.show()
```
